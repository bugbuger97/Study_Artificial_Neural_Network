# 한국어 및 영어의 언어 형태학적 차이
- 자연 언어의 3가지의 언어 형태
  - 고립어
    - 한자를 사용하는 중국어와 같이 단어의 형태가 시제, 인칭 변화 등에 따라 바뀌지 않고 고정된 형태의 언어
    - 굴절어
      - 영어, 독일어 등과 같이 시제, 인칭 변화 등에 따라 단어의 형태가 변하는 언어
    - 교착어
      - 한국어, 일본어 등과 같이 단어에 어미가 붙어 시제, 인칭 등의 뜻이 최종 결정되는 언어

# 한국어 토큰화의 어려움
- 한국어 텍스트 토큰화의 어려움
  - 띄어쓰기로 split을 수행하여 토큰화를 하였을 때, 단어와 조사가 붙어있게 되므로, 같은 의미라도 어미로 붙는 조사가 다르면, 서로 다른 단어로 인식하게 됨.
  
  - 한국어의 경우, 띄어쓰기 맞춤법이 있긴 하지만, 띄어쓰기가 잘 지켜지지 않는 corpus도 많으므로 기본적인 split방식으로는 만족스러운 토큰들을 얻기 어려움.

  - 한국어 텍스트에서 영어와 유사한 토큰들을 얻기 위해서는 형태소 단위의 분리가 필요하다.

# 형태소 분석기
- 형태소의 정의
  - 형태소(morpheme)란 의미를 가진 가장 작은 단위의 말을 의미함.

- 형태소 종류
  - 자립 형태소
    - 접사, 어미, 조사등과 상관없이 자립하여 사용할 수 있는 형태소
  
  - 의존 형태소
    - 다른 형태소와 결합하여 사용되는 형태소이며 접사, 어미, 조사 등이 이에 해당함.

- 형태소 분석기
  - 원문 텍스트에서 형태소를 추출하여 토큰화 해주는 모듈 혹은 프로그램을 형태소 분석기하고 함.

  - 한국어 NLP 파이썬 라이브러리인 KoNLPy가 가장 널리 쓰이는 형태소 분석기 라이브러리이며, 그 외에도 카카오에서 공개한 딥러닝 기반 형태소 분석기 khaii, 전자 통신 연구원 주도의 대규모 언어 지능 프로젝트 엑소브레인의 형태소 분석기 등도 있다.

  - KoNLPy의 메서드 분석
    - morphs 메서드
      - 문장을 입력하여 형태소 분석이 가능
    - pos 메서드
      - 문장에 대해 형태소 분석 및 품사 태깅을 한다.
      - 품사 태깅이란, 각 형태소마다 품사를 예측한 후 형태소-품사 pair를 만드는 것을 의미함.

# 사전학습 언어 모델
- 언어 모델
  - 언어 모델(Language Model, LM)은 주어진 문장 이후에 올 단어의 확률을 예측하는 모델을 의미
    - 확률을 예측할 때는 모든 단어에 대해서 확률을 매기며, 가장 확률이 높은 단어가 모델이 예측하는 다음 단어가 됨.
  
  - 주어진 문장을 보고 그 다음에 올 가장 그럴듯한 단어를 예측하는 모델이라고 볼 수 있음.

  - 잘 학습된 언어 모델만으로도 언어 모델에서 예측한 다음 단어로 새로운 sequence를 생성하고, 또 다음 단어를 예측하는 것을 반복하며 새로운 문장을 생성해낼 수 있음.

- 사전학습 언어 모델
  - 언어 모델을 학습시킬 때, corpus만 있으면 기계적으로 시퀀스 및 다음 단어를 만들어서 학습 데이터로 활용 가능함. 즉, 별도의 수작업을 통한 라벨링이 필요하지 않음.

  - 대규모의 데이터셋에 대해 수행하기 좋은 특성이 있으며, 또한 언어 모델이라는 것은 문맥을 어느 정도 고려하여 다음 단어를 예측하여야 하므로 해당 언어에 관한 전반적인 지식이 축적된 모델이라고 볼 수 있음.

  - 연구자들은 컴퓨터 비전에서의 전이학습처럼 자연어 처리에서도 언어 모델을 다른 세부적인 작업을 수행하기 위한 모델의 가중치 초기화 용도로 사용하기 시작하였으며, 이러한 목적으로 사용되는 언어 모델을 사전학습 언어모델(Pre-trained LM)이라고 한다.

  - OpenAI에서는 자연어 및 시퀀스 처리에 강점을 보이는 Transformer 모델을 활용한 사전학습 언어 모델을 선보임.
    - GPT(Generative Pre-trained Transformer)로 불리는 이 모델은 현재 점점 더 방대한 데이터셋에 대해 학습하여 GPT-4(& GPT-4 Turbo) 버전까지 공개

# BERT를 이용한 네이버 영화 리뷰 감성 분석
- BERT
  - BERT(Bidirectional Encoder Representations from Transfomers)는 2018년에 구글이 공개한 사전 학습 언어 모델로써 기존의 언어모델과는 약간 다른 방식의 언어모델 학습 방식을 갖고 있음.

  - Transformer를 활용한 모델 구조를 갖고 있으며, 다양한 task에 대한 전이 학습의 성능이 매우 좋아 주목받게 되었고 현재까지도 후속 모델들이 널리 활용되고 있음.

  - GPT가 스케일을 키운 이후로 더욱 압도적인 성능이 나오게 되며, 여러가지 후속 연구가 진행되었고, 현재까지는 스케일 = 성능이라는 공식이 거의 성립하게 됨. 최근 자연어 처리 분야에서는 주요 거대 기업들은 가능하면, 더욱 큰 스케일의 언어 모델을 개발하려고 경쟁중이며 이러한 거대 모델들을 LLM(Large Language Model)이라고 함.

- BERT 사전학습
  - BERT는 사전학습 진행 시, 기존 언어모델들이 단방향으로만 예측을 진행하는 것을 보완하기 위해 양방향에 대한 분석을 수행할 수 있도록 랜덤으로 단어를 지우고(마스킹) 해당 단어를 주변 문맥을 통해 예측하도록 하였음.
    - 이와 같은 형태의 언어모델을 maked language model(MLM)이라고 함.
    - ex) dog라는 단어가 지워진 상태로 예측하도록 학습시켰을 때, BERT에서 cat이라고 예측한 상황을 나타내고 있음.

- BERT fine-tuning
  - 사전 학습된 모델을 실제 세부 문제에 활용하는 단계를 fine-tuning이라고 함.
  
  - 텍스트 분류 fine-tuning 구조
    - Transformer의 출력값들 중 첫번째 토큰에 전결합층을 연결 후 분류 수행함.
  
  - 사전 학습 및 fine-tuning 모두에서 BERT의 입력값 중 맨 첫번째 토큰은 분류를 목적으로 한 특수 토큰인 [CLS]가 사용됨.

- 텍스트 태깅(tagging) fine-tuning 구조
  - 각 단어의 품사 등을 태깅하는 작업의 경우 각 단어의 index에 해당하는 출력 토큰들에 각각 전결합층을 연결해 태깅을 수행함.

- 텍스트 쌍(pair)에 대한 분류 fine-tuning 구조
  - BERT에서는 [SEP] 토큰을 통해 각 문장을 구분하며, 두 문장을 입력하고 분류를 수행할 수도 있음.(이미 사전학습 때도 NSP를 수행 시 문장을 연결해서 입력하였음.)

- BERT를 이용한 네이버 영화 리뷰 감성 분석 코드 과정
  - 1. BERT를 활용하기 위해 Transformers 설치 및 데이터셋 로드
  - 2.테스트 데이터를 전처리 해주지 않으면, 토큰화 과정에서 에러가 발생하기 때문에,  데이터와 같은 형식으로 일괄 전처리를 진행함.
  - 3. Trainer에 입력해주기 위한 데이터셋 생성
  - 4. 학습 환경 설정 및 정확도 확인을 위한 별도 함수 작성
  - 5. Trainer 객체 생성 및 학습 진행
  - 6. 모델 성능 평가 진행

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import urllib.request
from konlpy.tag import Okt
from tqdm import tqdm
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import TensorDataset, DataLoader
from collections import Counter
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt", filename="ratings_train.txt")
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt", filename="ratings_test.txt")
```

```python
train_data = pd.read_table('ratings_train.txt')
test_data = pd.read_table('ratings_test.txt')
```

```python
print ('Train data len :',len(train_data))
train_data[:5]
```

```python
print ('Test data len :',len(test_data))
test_data[:5]
```

```python
train_data.drop_duplicates(subset=['document'], inplace=True) # remove dulicates
```

```python
train_data.groupby('label').size()
```

```python
#한글 및 공백 이외 제거
train_data['document'] = train_data['document'].str.replace("[^ㄱ-ㅎ ㅏ-ㅣ 가-힣]", "", regex=True) # 마지막에 regex=True 추가
train_data[:5]

# regex=True: 문자열을 정규 표현식으로 해석
# 즉, 패턴을 복잡한 규칙으로 정의할 수 있으며, 특정 조건에 맞는 문자열을 찾아 변환하거나 제거할 수 있음
# regex=False: 문자열을 단순한 리터럴 텍스트(문자 그대로의 값)로 처리
# 정규 표현식으로 해석하지 않고, 입력된 패턴을 단순한 텍스트로 인식하여 처리
```

```python
# 공백 제거
train_data['document'] = train_data['document'].str.replace("^ +", "", regex=True) # 마지막에 regex=True 추가
train_data['document'] = train_data['document'].replace('', np.nan) # 강의자료: train_data['document'].replace('', np.nan, inplace=True) -> 지금 코드로 변경 추천
print(train_data.isnull().sum())
```

```python
train_data.loc[train_data.document.isnull()][:5]
```

```python
#null 제거, 하나라도 NaN이 포함된 경우 해당 행을 삭제
train_data = train_data.dropna(how = 'any')
print(len(train_data))
```

```python
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```python
okt = Okt()
X_train = []
for sentence in tqdm(train_data['document']):
  tokenized_sentence = okt.morphs(sentence, stem=True) # 토큰화
  stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] # 불용어 제거
  X_train.append(stopwords_removed_sentence)
print(X_train[:3])
```

```python
X_test = []
for sentence in tqdm(test_data['document']):
  tokenized_sentence = okt.morphs(str(sentence), stem=True) # 토큰화, string으로 캐스팅
  stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] # 불용어 제거
  X_test.append(stopwords_removed_sentence)
```

```python
#토큰화 된 단어들에 대해서 vocab 생성 및 단어 인덱스 형태의 데이터셋 생성
from collections import Counter

def tokenize(x_train,y_train,x_val,y_val):
  word_list = []

  for sent in x_train:
    for word in sent:
      word_list.append(word)

  corpus = Counter(word_list)
  # sorting on the basis of most common words
  corpus_ = sorted(corpus,key=corpus.get,reverse=True)[:10000]
  # creating a dict
  onehot_dict = {w:i+1 for i,w in enumerate(corpus_)}

  # tokenize
  final_list_train,final_list_test = [],[]
  for sent in x_train:
    final_list_train.append([onehot_dict[word] for word in sent
                                    if word in onehot_dict.keys()])
  for sent in x_val:
    final_list_test.append([onehot_dict[word] for word in sent
                                  if word in onehot_dict.keys()])

  return np.array(final_list_train, dtype=object), np.array(y_train),np.array(final_list_test, dtype=object), np.array(y_val),onehot_dict
```

```python
def padding_(sentences, seq_len):
  features = np.zeros((len(sentences), seq_len),dtype=int)
  for ii, review in enumerate(sentences):
    if len(review) != 0:
      features[ii, -len(review):] = np.array(review)[:seq_len]
  return features

x_train_pad = padding_(x_train,50)
x_test_pad = padding_(x_test,50)
```

```python
# create Tensor datasets
train_data = TensorDataset(torch.from_numpy(x_train_pad), torch.from_numpy(y_train))
test_data = TensorDataset(torch.from_numpy(x_test_pad), torch.from_numpy(y_test))

# dataloaders
batch_size = 50

# make sure to SHUFFLE your data
train_loader = DataLoader(train_data, shuffle=True, batch_size=batch_size)
test_loader = DataLoader(test_data, shuffle=False, batch_size=batch_size)
```

```python
class GRU_model(nn.Module):
  def __init__(self, n_layers, hidden_dim, n_vocab, embed_dim, n_classes, device):
    super(GRU_model, self).__init__()
    self.n_layers = n_layers
    self.hidden_dim = hidden_dim
    self.device = device

    self.embed = nn.Embedding(n_vocab, embed_dim)
    self.gru = nn.GRU(embed_dim, self.hidden_dim,
                      num_layers=self.n_layers,
                      batch_first=True)
    self.out = nn.Linear(self.hidden_dim, n_classes)

  def forward(self, x):
    x = self.embed(x)
    h_0 = self._init_state(batch_size=x.size(0)) # 첫번째 히든 스테이트를 0벡터로 초기화
    x, _ = self.gru(x, h_0)  # GRU의 리턴값은 (배치 크기, 시퀀스 길이, 은닉 상태의 크기)
    h_t = x[:,-1,:] # (배치 크기, 은닉 상태의 크기)의 텐서로 크기가 변경됨. 즉, 마지막 time-step의 은닉 상태만 가져온다.
    logit = self.out(h_t)  # (배치 크기, 은닉 상태의 크기) -> (배치 크기, 출력층의 크기)
    return logit

  def _init_state(self, batch_size):
    new_state = torch.zeros(self.n_layers, batch_size, self.hidden_dim).to(self.device)
    return new_state
```

```python
is_cuda = torch.cuda.is_available()

# If we have a GPU available, we'll set our device to GPU. We'll use this device variable later in our code.
if is_cuda:
  device = torch.device("cuda")
  print("GPU is available")
else:
  device = torch.device("cpu")
  print("GPU not available, CPU used")
```

```python
n_layers = 1
vocab_size = len(vocab) + 1  # extra 1 for <pad>
hidden_dim = 128
embed_dim = 100
n_classes = 2

model = GRU_model(n_layers, hidden_dim, vocab_size, embed_dim, n_classes, device).to(device)
```

```python
def train(model, criterion, optimizer, data_loader):
  model.train()
  train_loss = 0
  for i, (x, y) in enumerate(data_loader):
    x, y = x.to(device), y.to(device)

    optimizer.zero_grad()
    logit = model(x)
    loss = criterion(logit, y)
    loss.backward()
    optimizer.step()

    train_loss += loss.item() * x.size(0)

  return train_loss / len(data_loader.dataset)

def evaluate(model, data_loader):
  model.eval()
  corrects, total_loss = 0, 0
  for i, (x, y) in enumerate(data_loader):
    x, y = x.to(device), y.to(device)

    logit = model(x)
    corrects += (logit.max(1)[1].view(y.size()).data == y.data).sum()
  size = len(data_loader.dataset)

  avg_accuracy = 100.0 * corrects / size
  return avg_accuracy
```

```python
num_epochs = 10
lr = 0.001

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)

for e in range(1, num_epochs+1):
  train_loss = train(model, criterion, optimizer, train_loader)
  test_accuracy = evaluate(model, test_loader)

  print("[Epoch: %d] train loss : %5.2f | test accuracy : %5.2f" % (e, train_loss, test_accuracy))
```

```python
import pandas as pd
import re
import urllib.request
import numpy as np

urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt", filename="ratings_train.txt")
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt", filename="ratings_test.txt")
```

```python
train_data = pd.read_table('ratings_train.txt')
test_data = pd.read_table('ratings_test.txt')
```

```python
# 전처리 진행
train_data.drop_duplicates(subset=['document'], inplace=True) # 중복 데이터 제거
train_data['document'] = train_data['document'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","") # 한글 및 공백 이외 제거
train_data['document'] = train_data['document'].str.replace('^ +', "") # white space 데이터를 empty value로 변경
train_data['document'].replace('', np.nan, inplace=True)
train_data = train_data.dropna(how = 'any') # null 값 제거
```

```python
test_data['document'] = test_data['document'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","", regex=True)  # 한글 및 공백 이외 제거, 마지막에 regex=True 추가
test_data['document'] = test_data['document'].str.replace('^ +', "", regex=True) # white space 데이터를 empty value로 변경, 마지막에 regex=True 추가
train_data['document'] = train_data['document'].replace('', np.nan)  # 공백을 nan으로 대체
test_data = test_data.dropna(how = 'any')  # null값 제거
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

model = AutoModelForSequenceClassification.from_pretrained("klue/bert-base", num_labels=2)
tokenizer = AutoTokenizer.from_pretrained("klue/bert-base")
```

```python
sampled_train_data = train_data[:10000]
sampled_test_data = test_data[:10000]

train_tokens = tokenizer(list(sampled_train_data['document']), padding='max_length', truncation=True, return_tensors="pt", add_special_tokens=True)
test_tokens = tokenizer(list(sampled_test_data['document']), padding='max_length', truncation=True, return_tensors="pt", add_special_tokens=True)

print(tokenizer.convert_ids_to_tokens(train_tokens['input_ids'][0]))
```

```python
import torch

class BERTDataset(torch.utils.data.Dataset):
  def __init__(self, encodings, labels):
    self.encodings = encodings # 토큰화된 데이터
    self.labels = labels # 레이블 데이터

  def __getitem__(self, idx):
    item = {key: val[idx] for key, val in self.encodings.items()}  # 각 샘플 데이터
    item['labels'] = torch.tensor(self.labels[idx]) # 레이블 추가
    return item

  def __len__(self):
    return len(self.labels) # 데이터 크기 반환

train_dataset = BERTDataset(train_tokens, list(sampled_train_data['label']))
test_dataset = BERTDataset(test_tokens, list(sampled_test_data['label']))
```

```python
from transformers import Trainer, TrainingArguments
from sklearn.metrics import accuracy_score

training_args = TrainingArguments(
    output_dir='./results',          # 학습 결과 저장 디렉토리
    num_train_epochs=1,              # 총 학습 에포크 수
    per_device_train_batch_size=10,  # 학습 시 디바이스당 배치 크기
    per_device_eval_batch_size=20,   # 평가 시 디바이스당 배치 크기
    warmup_steps=100,                # 학습률 스케줄링을 위한 워밍업 스텝 수
    weight_decay=0.01,               # 가중치 감쇠(L2 정규화) 값
    logging_dir='./logs',            # 로그 저장 디렉토리
    logging_steps=250,               # 몇 스텝마다 로그를 기록할지 설정
    report_to="none",                # WandB 또는 다른 로깅 도구를 사용하지 않음
)

def compute_metrics(pred):
  labels = pred.label_ids              # 실제 레이블
  preds = pred.predictions.argmax(-1)  # 모델 예측값 (가장 높은 확률 선택)

  acc = accuracy_score(labels, preds)  # 정확도 계산
  return {
      'accuracy': acc                  # 정확도를 반환
  }

trainer = Trainer(
    model=model,                         # 학습할 모델 (예: BERT)
    args=training_args,                  # 학습 설정
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset,           # evaluation dataset
    compute_metrics=compute_metrics      # additional evaluation metrics
)

trainer.train()
```
